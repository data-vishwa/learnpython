### Speed

```python
from IPython.display import Image
Image("images/array_vs_list.png")
```

```python
python_list_1 = range(1,1000)
python_list_2 = range(1,1000)


numpy_list_1 = np.arange(1,1000)
numpy_list_2 = np.arange(1,1000)
```

```python
%%capture timeit_python
%%timeit

[x + y for x, y in zip(python_list_1,python_list_2)]
[x - y for x, y in zip(python_list_1,python_list_2)]
[x * y for x, y in zip(python_list_1,python_list_2)]
[x / y for x, y in zip(python_list_1,python_list_2)]
```

```python
print(timeit_python)
```

```python
%%capture timeit_python
%%timeit

numpy_list_1 + numpy_list_2
numpy_list_1 - numpy_list_2
numpy_list_1 * numpy_list_2
numpy_list_1 / numpy_list_2
```

```python
print(timeit_python)
```

## Other Operations

```python
a = np.array([1, 1, 1, 3])
b = np.array([4, 2, 2, 4])
```

```python
a < b
```

```python
np.array_equal(a,b)
```

```python
a = np.array([1, 1, 0, 0], dtype=bool)
b = np.array([1, 0, 1, 0], dtype=bool)
```
